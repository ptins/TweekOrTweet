```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
```

```python
df = pd.read_csv('user_tweets_from.csv', header=None)
df.columns = ['tweet_id','created_at','favorite_count','retweet_count','followers_count','screen_name','industry']
df.set_index('tweet_id', inplace=True)

# feature processing
df['RFR'] = (df['favorite_count']+df['retweet_count'])/df['followers_count']
```

```python
df.head()
```

```python
# normalization
def normalize(df):
    result = df.copy()
    for feature_name in df.columns:
        print(type(df[feature_name].iloc[0]))
        if (type(df[feature_name].iloc[0]) == numpy.int) | (type(df[feature_name].iloc[0]) == numpy.float):
            max_value = df[feature_name].max()
            min_value = df[feature_name].min()
            result[feature_name] = (df[feature_name] - min_value) / (max_value - min_value)
    return result
```

```python
t = normalize(df)
```
